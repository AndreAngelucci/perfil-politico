```python
import pandas as pd
```

```python
tweets = pd.read_csv('../data/_dirty/tweets.csv')
```

```python
tweets.head()
```

```python
deputy_account = pd.read_csv('../data/_serenata/2018-02-05-congresspeople-social-accounts.xz')
deputy_account.head()
```

```python
df = pd.merge(tweets, deputy_account, how='left', left_on='usernameTweet', right_on='twitter_profile')
```

```python
df.head()
```

```python
df.congressperson_id.fillna(0, inplace=True)
df.congressperson_id = df.congressperson_id.astype(int)
```

```python
df.text = df.text.apply(lambda x: f'"{x}"')
```

```python
df.head()
```

```python
df.to_csv('../data/tweets_with_ids_new.csv', index=False, sep=';')
```
