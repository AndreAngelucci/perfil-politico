```python
import json
import pandas as pd
```

```python
df = pd.read_csv('../data/_results/asset_evolution_elections_2018.csv')
df = df[df.year.isin([2010, 2014, 2018])]
```

```python
counts = df.name.value_counts()
candidates = counts[counts>=2].index
```

```python
df_candidates = df[df.name.isin(candidates)]
df_candidates.head(2)
```

```python
df_candidates.shape
```

```python
df_candidates.sum_of_assets_declared = df_candidates.sum_of_assets_declared.str.replace('None', '0')
```

```python
df_candidates.fillna(0, inplace=True)
df_candidates.sum_of_assets_declared = df_candidates.sum_of_assets_declared.astype(float)
```

```python
df_pivot = df_candidates.pivot_table(columns=['year'], index=['name'], values=['sum_of_assets_declared'])
df_pivot.fillna(0, inplace=True)
```

```python
# Retirando multiindex do resultado do pivot table (simplificando)

new_dict = {
    2010: df_pivot['sum_of_assets_declared'][2010],
    2014: df_pivot['sum_of_assets_declared'][2014],
    2018: df_pivot['sum_of_assets_declared'][2018],
}

df = pd.DataFrame(new_dict)
```

```python
df['2010-2014'] = df[2014] - df[2010]
df['2014-2018'] = df[2018] - df[2014]
df['2010-2018'] = df[2018] - df[2010]

# Transformando para milhões
for col in df.columns:
    df[col] = df[col] / 10**6
```

# Crescimento de patrimônio de 2010 a 2018

```python
df_2010_2018 = df[df['2010-2018']>=50]
```

```python
df_2010_2018.sort_values(by='2010-2018', ascending=False)
```

```python
df_2014_2018 = df[df['2014-2018']>=50]
df_2014_2018.sort_values(by='2010-2018', ascending=False)
```

```python
df_acima_50mi = df[(df[2010]>=50) | (df[2014]>=50) | (df[2018]>=50)]
df_acima_50mi.sort_values(by='2010-2018', ascending=False)
```
