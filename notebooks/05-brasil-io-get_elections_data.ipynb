```python
import json

import pandas as pd
import requests

from utils import get_collection
```

```python
collection = get_collection()
```

```python
genders = {'MASCULINO': 'male', 'FEMININO': 'female'}
```

```python
def get_person_basic_info(row):
    return {
        'cpf': row['cpf_candidato'],
        'gender': genders[row['descricao_sexo']],
        'state': row['sigla_uf_nascimento'],
        'civil_name': row['nome_candidato'],
        'birthday': row['data_nascimento'],
        
    }
```

```python
def get_election_info(row):
    return {
        'year': int(row['ano_eleicao']),
        'party': row['sigla_partido'],
        'elected': row['desc_sit_tot_turno'] == 'ELEITO',
        'position': row['descricao_cargo']
    }
```

```python
def add_elections_df_to_db(row):
    cpf = row['cpf_candidato']
    person = collection.find_one({'cpf': cpf})

    if not person:
        collection.insert_one(get_person_basic_info(row))
        person = collection.find_one({'cpf': cpf})

    person.update(get_person_basic_info(row))
    elections = person.get('elections')

    election_info = get_election_info(row)
    if elections and election_info not in elections:
        person['elections'].append(election_info)
    else:
        person['elections'] = [get_election_info(row)]

    collection.find_one_and_replace({'cpf': cpf}, person)
    print('Added {}'.format(cpf))
```

```python
url = 'https://brasil.io/api/dataset/eleicoes-brasil/data'
find_next = True
count = 0
dfs = []

while find_next:
    response = requests.get(url)
    data = json.loads(response.content)
    df = pd.DataFrame(data.get('results'))
    
    for i, row in df.iterrows():
        add_elections_df_to_db(row)
    
    find_next = data.get('next')
    count += 1
```
